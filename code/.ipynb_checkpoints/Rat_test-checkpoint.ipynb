```python
def Ratlist(filename,open):
    file_name = filename
    f = open(file_name,'r')
    final = []
    counter = 0
    for i in f.readlines():
    
        if counter>=4 and counter<=93:
            final.append(i.split()[2:22])
            if counter==93:
                break
        counter+=1
    return final
```

```python
counter = 0
Rat_matrix= []
label_name = []
check_name_list = []
import glob
for filename in glob.glob('*.smp'):

    Temp = Ratlist(filename,open)
    if len(Temp)<90:
        
        check_name_list.append(filename)
    else:
        label_name.append(filename)
        Rat_matrix.append(Temp)

    counter+=1
    

length_matrix=len(Rat_matrix)
```
